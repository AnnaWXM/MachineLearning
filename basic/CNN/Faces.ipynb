```python
# Face recognition with MTCNN 
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
from mtcnn import MTCNN
```

```python
# Load default MTCNN model and dtetct faces in the image with the model
model = MTCNN()
faces = model.detect_faces(pixels)

# First, add the original image at the bottom layer
# Then define the drawing context for the frames
plt.imshow(pixels)
ax = plt.gca()

# Print data for all recognised faces
for face in faces:
    print(face)
    # Get coordinates and dimensions for the face box
    x, y, width, height = face['box']
    #ax.imshow(pixels)
    # Create red rectangle around the face
    rect = Rectangle((x, y), width, height, fill=False, color='red')
    # Add the rectangle into the image
    ax.add_patch(rect)

    # Add title to the image
    plt.title('There are ' + str(len(faces)) + ' people in this picture')

    # Remove axis from the image
    plt.axis('off')

    # Save image as 'faces.png'
    plt.savefig('faces.png', bbox_inches='tight', pad_inches=0)

# Show the picture
plt.show()
```

```python
# For each face detected
for i in range(len(faces)):
    # Get coordinates and dimensions for the face box
    x1, y1, width, height = faces[i]['box']
    # Calculate all corner point coordinates of the face in the original image
    x2, y2 = x1 + width, y1 + height
    # Create a subplot of the face
    plt.subplot(1, len(faces), i + 1)
    # Clean axis
    plt.axis('off')
    # Show the face image
    plt.imshow(pixels[y1:y2, x1:x2])
    # Save the face image into a file with order number to separate filenames
    plt.savefig('face-' + str(i + 1) +'.png', bbox_inches='tight', pad_inches=0)
    # Show image and clean pyplot object before next face image
    plt.show()
    plt.close()
```
